```python
db = pd.read_csv('Extraction_UK_2018_07.csv', delimiter = '\t', header=0, encoding = 'utf-8',error_bad_lines=False, dtype = {'Zip Code': str, 'City':str} )
```

```python
db = db[['Vehicle_ID', 'Autobiz ID', 'Doublon_id', 'Autotrader ID',
       'Dealer Name', 'Make', 'Model', 'Version', 'ADS_VO_doublon_id','ADS_vo_REFERENCE_VERSION_ID','Market_value', 'Year',
       'Price', 'Mileage', 'Garage_id','Version','CARBURANT', 'Doublon_id','Vo_site_id', 'SITE', 'Main Activity', 'Adresse 1', 'Adresse 2',
       'Zip Code', 'City', 'Phone', 'Fax', 'Website', 'Site']]
```

```python
db = db.drop(['Doublon ID', 'Model', 'Version', 'ADS_VO_doublon_id', 'Mileage', 'Version', 'CARBURANT', 'Doublon_id', 'Website', 'Site'], axis=1)
```

```python
db = db.reset_index()
```

```python
db = db.rename(columns={'index':'Vehicle ID', 'Vehicle_ID':'Autobiz ID', 'Autobiz ID':'Autotrader ID', 'Autotrader ID':'Dealer Name', 'Dealer Name':'Make', 'Make':'Model', 'ADS_vo_REFERENCE_VERSION_ID':'Market Value', 'Market_value':'Year', 'Year':'Price', 'Price':'Mileage', 'Garage_id':'Version', 'Vo_site_id':'Energy', 'SITE':'Main activity', 'Main Activity':'Adress', 'Adresse 1':'Zipcode', 'Adresse 2':'City', 'Zip Code':'Phone', 'City':'Fax', 'Phone':'Website', 'Fax':'Marketplace'})
```

```python
db['Year'] = db['Year'].astype(str).apply(lambda x: x.replace('.0',''))
db['Price'] = db['Price'].astype(str).apply(lambda x: x.replace('.0',''))
```

```python
db
```

```python
db = db[db['Autobiz ID'] != 74641]
```

```python
db = db[db['Vehicle ID'] != '']
db = db[db['Zipcode'] != '']
```

```python
db['Marketplace'] = db['Marketplace'].str.replace('nan|[{}\s]','').str.split(',').apply(set).str.join(',').str.strip(',').str.replace(",{2,}",",")
```

```python
db['Energy'] = db['Energy'].replace(['n/a'], 'NOT SUPPLIED')
db['Energy'] = db['Energy'].replace(['hybrid-??-petrol/electric-plug-in'], 'Hybrid  Petrol/Electric Plug-in')
db['Energy'] = db['Energy'].replace(['Petrol/Electricity'], 'Hybrid  Petrol/Electric')
db['Energy'] = db['Energy'].replace(['PETROL/ELECTRIC'], 'Hybrid  Petrol/Electric')
db['Energy'] = db['Energy'].replace(['P/Electric'], 'Hybrid  Petrol/Electric')
db['Energy'] = db['Energy'].replace(['Electrique'], 'Electric')
db['Energy'] = db['Energy'].replace([''], 'NOT SUPPLIED')
db['Energy'] = db['Energy'].replace(['petrol-hybrid'], 'Hybrid  Petrol/Electric')
db['Energy'] = db['Energy'].replace(['Petrol Hybrid'], 'Hybrid  Petrol/Electric')
db['Energy'] = db['Energy'].replace(['Petrol/Electric Hybrid'], 'Hybrid  Petrol/Electric')
db['Energy'] = db['Energy'].replace(['Petrol Plug-in Hybri'], 'Hybrid  Petrol/Electric Plug-in')
```

```python
db['Zipcode'] = db['Zipcode'].replace([' '], '')
db['Zipcode'] = db['Zipcode'].replace(['&#8203;'], '')
```

```python
db['Main activity'] = db['Main activity'].replace(['INDEPENDANTS'], 'INDEPENDENT')
```

```python
db = db.drop('Version', 1)
```

```python
db[['Dealer Name', 'Make', 'Model', 'Main activity', 'Energy', 'Adress', 'Zipcode', 'City']] = db[['Dealer Name', 'Make', 'Model', 'Main activity', 'Energy', 'Adress', 'Zipcode', 'City']].apply(lambda x: x.str.upper())
```

```python
db['Marketplace'] = db['Marketplace'].str.split(',').map(lambda x: ','.join(sorted(x)))
```

```python
db.to_csv(database_extraction_current_month_after_script, index = False)
```
